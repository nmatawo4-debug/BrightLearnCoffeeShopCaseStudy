```sql
select * 
from `workspace`.`default`.`bright_coffee_shop_case_study`;



SELECT min(transaction_time) AS min_time,
       max(transaction_time) AS max_date
from `workspace`.`default`.`bright_coffee_shop_case_study`;
     


---min date of data collection - 01/01/2023
SELECT min(transaction_date) AS min_date
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


----max date of data collection - 30/06/2026 (period - 6 months)
SELECT max(transaction_date) AS max_date
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


--- stores (Lower Manhattan, Hell's Kitchen, Astoria)
SELECT DISTINCT store_location 
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


SELECT COUNT(DISTINCT store_location) AS num_of_stores
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


--- distinct products - 29 products
SELECT COUNT (DISTINCT product_type)
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


-------------------------------------------------
-- 1. Checking product prices
------------------------------------------------
SELECT MIN(unit_price) As cheapest_price,
       MAX(unit_price) As expensive_price
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


-------------------------------------------------
-- 3. Checking products sold at our stores 
------------------------------------------------
SELECT DISTINCT product_category
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;



SELECT DISTINCT product_detail
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


SELECT DISTINCT product_type
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


SELECT DISTINCT product_category AS category,
                product_detail AS product_name
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


-------------------------------------------------
--- 
------------------------------------------------

SELECT 
COUNT(*) AS number_of_rows,
      COUNT(DISTINCT transaction_id) AS number_of_sales,
      COUNT(DISTINCT product_id) AS number_of_products,
      COUNT(DISTINCT store_id) AS number_of_stores
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;

------------------------------------------------


SELECT transaction_id,
      transaction_date,
      Dayname(transaction_date) AS Day_name,
      Monthname(transaction_date) AS Month_name,
      transaction_qty*unit_price AS revenue_per_tnx
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;

-----------------------------------------------------



SELECT 
    COUNT (*) AS num_of_rows,
 COUNT(DISTINCT product_id) AS num_product_id,
       COUNT(DISTINCT product_type) AS num_producttype,
       COUNT(DISTINCT product_category) AS num_category
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


---transaction day, month, revenue
SELECT transaction_date,
      dayname(transaction_date) AS day,
      monthname(transaction_date) AS month,
      SUM(transaction_qty*unit_price) AS revenue
FROM `workspace`.`default`.`bright_coffee_shop_case_study`
GROUP BY  transaction_date,
            day,
            month;


---dates
SELECT transaction_time,
      dayname(transaction_date) AS day_name,
      monthname (transaction_date) AS month_name,
    CASE 
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '08:59:59' THEN 'rush hour'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '09:00:00' AND '11:59:59' THEN 'morning'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '15:59:59' THEN 'afternoon'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '16:00:00' AND '23:59:59' THEN 'rush hour'
      END AS time_buckets,
  
   CASE 
      WHEN day_name IN ('Sat', 'Sun') THEN 'weekend'
      ELSE 'weekday'
   END AS day_classification,

---Count ids
      COUNT(DISTINCT store_id) AS num_of_stores,
      COUNT(DISTINCT product_id) AS num_products,
      COUNT(DISTINCT transaction_id) AS num_of_sales,

-----revenue
     SUM(transaction_qty*unit_price) AS revenue,

---Categorical Column
   store_location,
   product_category,
   product_detail
FROM `workspace`.`default`.`bright_coffee_shop_case_study`
GROUP BY transaction_date,
            transaction_time,
          day_name,
          month_name,
          store_location,
          product_category,
          product_detail,
          time_buckets,
          day_classification;




SELECT *
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;
```
